```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

# Filter the data from year 2000 through the most recent year
recent_data = df[df['Year'] >= 2000]

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'], color='blue', marker='o', alpha=0.7, label='Data')

# Calculate the line of best fit using linregress for the recent data
slope_recent, intercept_recent, _, _, _ = linregress(recent_data['Year'], recent_data['CSIRO Adjusted Sea Level'])

# Plot the line of best fit for the recent data
plt.plot(recent_data['Year'], slope_recent * recent_data['Year'] + intercept_recent, color='red', label='Line of Best Fit')

# Extend the line to the year 2050
future_years = range(2000, 2051)
plt.plot(future_years, slope_recent * future_years + intercept_recent, linestyle='--', color='green', label='Prediction for 2050')

# Add title and labels
plt.title('Rise in Sea Level')
plt.xlabel('Year')
plt.ylabel(' Sea Level (inches))')

# Show the legend
plt.legend()

# Show the plot
plt.grid(True)  # Add grid for better readability
plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()

# Predict sea level rise in 2050
sea_level_2050_recent = slope_recent * 2050 + intercept_recent
print(f"Predicted sea level rise in 2050 (since 2000): {sea_level_2050_recent:.2f} mm")
```
